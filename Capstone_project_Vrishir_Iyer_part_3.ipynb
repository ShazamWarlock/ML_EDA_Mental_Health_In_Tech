#Exploratory Data Analysis of "Mental health in Technology" Survey Data
##About the dataset
This dataset is from a 2014 survey that measures attitudes towards mental health and frequency of mental health disorders in the tech workplace. Our dataset contains 1259 rows and 27 columns (features).

The dataset contains the following features:
1. Timestamp (Numerical) : This shows the time that the data was recorded.

2. Age (Numerical) : What is the age of the person? There could be some correlation between age and how people perceive or react to mental health illnesses.

3. Gender (Categorical): The gender of the person (male or female). Gender could play a role in determining the response of a person to mental health illnesses. For example, women may react and comprehend some mental health issues better than men and vice versa.

4. Country (Categorical) : What country do they belong to? Each country has its own set of cultures and while some may have cultures that are inclined more towards mental health management than others.

5. State (Catgeorical) : If they are from the US, then which state to they belong to in the US.

6. self_employed (Categorical) : Are they self-employed? Self-employment comes with its own set of advantages (like being your own boss) and disadvantages (insecurities, not having a stable income), which could have an effect on mental health.

7. family_history (Categorical) : Do they have a history of mental illness in their family? While most mental health illnesses are not genetically transfered, they are surely transfered during childhood. Thus, having data regarding the family history could be an important factor affecting mental health related issues.

8. treatment (Categorical) : Have they received treatment for a mental condition earlier? This may help in determining the likelihood of a person having a mental health condition.

9. work_interfere (Categorical) : Do they feel like their mental health condition affects their work, considering that they have one?

10. no_employees (Categorical) : How many employees does their company have? People tend to react differently to social environments depending on their size. Some are unaffected by the fact that they are merely a grain of sand in a field while some may feel insignificant. That is why we analyse this data to see how it affects mental health.

11. remote_work (Categorical): Do they work remotely at least 50% of the time? For some people, working remotely might be more relaxing since they are working from their comfort space.

12. tech_company (Categorical) : Is their company primarily a tech company/organisation?

13. benefits (Categorical): Does their employer provide mental health benefits? If the organisation is providing such benefits, it could be a saving grace for someone undergoing mental distress.

14. care_options (Categorical): Do they know the options for mental health that their employer provides? Often times, people are unaware of such options and remain in need of help when the help is just a few feet away from them. Thus lack of awareness is an important factor.

15. wellness_program (Categorical): Has their employer ever discussed mental health as part of their employee wellness program? Giving mental health a priority would lead to more people taking such help. One of the major reasons that people don't seek help regrading mental health illnesses is because they do not give it priority. Thus, it could be an important factor in our study.

16. seek_help (Categorical): Does their employer provide resources to learn more about mental health issues and seek help regarding the same?

17. anonymity (Categorical): Is their anonymity protected if they choose to take advantage of mental health or substance abuse treatment resources? Being anonymous gives a source of comfort for people, since it is in human nature as social creatures to feel and act on the basis of peer judgements. A person has to feel safe while being vulnerable, whether emotionally or physically. Thus, it is an important factor in our analysis.

18. leave (Categorical): How easily do they get medical leave for a mental health condition? This might be an indicator regarding how seriously the institution takes mental health on a more realistic scale.

19. mental_health_consequence (Categorical): Do they think that discussing a mental health issue with their employer would have negative consequences for them? If they do, then most of them would not come forward and address their issues, which is not an healthy scenario.

20. phys_health_consequence (Categorical): Do they think that discussing a physical health issue with their employer would have negative consequences?

21. coworkers (Categorical): Would they be willing to discuss a mental health issue with their coworkers? Discussions among peers can be very healthy. In general, people respond positively when they see someone going through something similar to them. A little bit of empathy fixes the feeling of loneliness, leading to positive results. Thus, this could be an important factor in our analysis.

22. supervisor (Categorical): Would they be willing to discuss a mental health issue with their direct supervisor? Transperancy is something that could both positively and negatively impact someone's condition. As a supervisor, one gets to work closely with that person and thus opens a gateway for them to understand and comprehend the person's mental condition better.

23. mental_health_interview (Categorical): Would they bring up a mental health issue with a potential employer in an interview?

24. phys_health_interview (Categorical): Would they bring up a physical health issue with a potential employer in an interview?

25. mental_vs_physical (Categorical): Do they feel that their employer takes mental health as seriously as physical health?

26. obs_consequence (Categorical): Have they heard of or observed negative consequences for coworkers with mental health conditions in their workplace? This might lead to people safeguarding their mental health conditions even more, thus leading to zero transperancy of their condition, which is not a healthy outcome. Thus, this could be an important factor in our study.

#Model Training
In this step we split the model into testing and training sets. This step allows the model to identify patterns and relationships between the feature variables.
In our particular case which is classification, it allows the model it allows the model to classify the treatment variable into categories.  
Another goal of training is to generalize the model to new unseen data well. Along with that this step also allows the model to adapt to complex relationships between features and test hypothesis.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
from sklearn.metrics import mean_squared_error, precision_recall_curve
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc, RocCurveDisplay

input_features = ['Age_Capped', 'Gender', 'family_history', 'benefits', 'care_options', 'anonymity', 'leave', 'work_interfere']
X = df[input_features]
y = df['treatment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

#Model(s) Evaluation
Here, we evaluate both the models using certain metrics to determine which model performs better.

##Random Forest Classifier

###Prediction
We fit the random forest classifier with our data and analyse the accuracy.
As we can see, the accuracy is around 71%, which means that it is close to being acceptable but isnt quite acceptable yet.

This is why I have proceeded to use cross-validation techniques to optimize the model and increase its accuracy.

```python
forest = RandomForestClassifier(n_estimators=250, random_state=0)       #initializing the model
forest.fit(X_train, y_train)      #fit the model using the trained data

y_pred = forest.predict(X_test)     #predict the treatment
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

###Cross Validation for Optimization
We print out the cross validation scores of the model. As we can see, the mean CV accuracy is around 71.8% which means that across all the folds of the model, this is the average accuracy. This means that there may be some variation of the model, with the correct parameters, that may give us a higher accuracy.

CV standard deviation being 1.51% shows the variation in each fold.




```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(forest, X, y, cv=5, scoring='accuracy')

print(f"Mean CV Accuracy: {np.mean(cv_scores) * 100:.2f}%")
print(f"CV Standard Deviation: {np.std(cv_scores) * 100:.2f}%")
```

###Hyperparamater Tuning
In this step, we select the best possible hyperparameters for our model. In Random Forest classifier, this is the number of trees and the depth of each tree.

We form a random grid with the specified range of hyperparameters and other variables. This grid is passed as an argument to RandomizedSearchCV.

RandomizedSearchCV will compare all the variations of the models and produce the best result out of all.

In this step we manage to find the best model using the best parameters having the best cross-validation accuracy.

```python
from sklearn.model_selection import RandomizedSearchCV

random_grid = {
    'n_estimators': [50, 100, 200, 300, 500],       #number of trees to find the optimal number of trees for classification
    'max_depth': [None, 10, 20, 30, 50],            #optimal depth of each tree to find the most optimal solution
    'min_samples_split': [2, 5, 10],                #the number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],                  #the minimum number of samples required to be at leaf node
    'bootstrap': [True, False]                      #
}

#this finds the best combination for random forest. It specifies the number of iterations and other parameters
random_search = RandomizedSearchCV(estimator=forest, param_distributions=random_grid, n_iter=100, cv=5, scoring='accuracy', random_state=0, verbose=2, n_jobs=-1)

random_search.fit(X, y)

print(f"Best Parameters: {random_search.best_params_}")
print(f"Best Cross-Validation Accuracy: {random_search.best_score_ * 100:.2f}%")
```

###Refit the model with Optimal Parameters
After finding out the best parameters in the previous step, we use those parameters and refit the model with them.

Our result, as we can see, has improved considerably.

There is a 5% increase in model accuracy after cross-validation.

```python
optimized_forest = RandomForestClassifier(**random_search.best_params_, random_state=0)       #we refit the model with the new parameters that we got from cross-validation
optimized_forest.fit(X_train, y_train)

y_pred = optimized_forest.predict(X_test)
y_pred_proba = optimized_forest.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
print(f"Optimized Model Accuracy: {accuracy * 100:.2f}%")
```

3. ROC Curve: The curve plots the trade-off between True Positive Rate (TPR) and False Positive Rate (FPR) at different decision thresholds. ```python

```python
importances = optimized_forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in optimized_forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

labels = [input_features[i] for i in indices]

#plotting the feature importances
plt.figure(figsize=(12, 8))
plt.title("Feature Importances RF")
plt.bar(range(X.shape[1]), importances[indices], color="#8DA1B9", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), labels, rotation='vertical')
plt.xlim([-1, X.shape[1]])
plt.ylabel("Importance Score")
plt.show()

#calculating the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

#printing the classification report which shows the values of f1-score, precision and recall
class_report = classification_report(y_test, y_pred)
print('Classification Report:')
print(class_report)

#displaying the confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Treatment', 'Treatment'], yticklabels=['No Treatment', 'Treatment'])
plt.xlabel('Predicted_values')
plt.ylabel('Actual_values')
plt.title('Confusion Matrix')
plt.show()

#specifying the ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

#plotting the figure for ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False_Positive_Rate')
plt.ylabel('True_Positive_Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
